
# Installing Jupyter, Starting JupyterLab, and Uploading a Notebook

## Step 1: Install Jupyter
First, you need to install Jupyter on your system. Open your command prompt or terminal and run the following command:

```bash
pip install jupyterlab --user
```

## Step 2: Start JupyterLab
Once the installation is complete, you can start JupyterLab. Run the following command in your command prompt or terminal:

```bash
C:\Users\Ceres\AppData\Roaming\Python\Python312\Scripts\jupyter-lab.exe
```

This command will start JupyterLab and open it in your default web browser at `http://localhost:8888`.

## Step 3: Uploading a Notebook File
Follow these steps to upload a Jupyter notebook file into your JupyterLab instance:

1. **Open JupyterLab:**
   - Navigate to `http://localhost:8888` in your web browser if it hasn't opened automatically.

2. **Navigate to the Home Directory:**
   - In the JupyterLab interface, you'll see a file browser panel on the left side.

3. **Upload the Notebook File:**
   - Click the "Upload Files" button (an upward-pointing arrow) in the file browser panel.
   - Browse to the location where your notebook file is saved (e.g., `Tweepy_Installation_and_Usage.ipynb`) and select it.
   - The notebook file will be uploaded to the current directory shown in the file browser.

4. **Open the Notebook File:**
   - Once the file is uploaded, you should see it listed in the file browser.
   - Double-click on the notebook file to open it in a new tab within JupyterLab.

## Step 4: Select the Kernel
When you open the notebook, JupyterLab may prompt you to select a kernel. A kernel is the execution environment for your code. To select the appropriate kernel:

1. **Choose the Kernel:**
   - When the "Select Kernel" dialog appears, choose `Python 3` or the kernel that corresponds to the Python environment where you installed Tweepy.

2. **Run the Notebook:**
   - Once the kernel is selected, you can start running the cells in the notebook.
   - To run all cells, go to "Run" > "Run All Cells" from the JupyterLab menu.
   - Alternatively, run each cell individually by clicking on the cell and pressing `Shift + Enter`.

## Additional Notes
- If you encounter any issues with kernel selection, you can list available kernels by opening a new terminal in JupyterLab and running:
  ```bash
  jupyter kernelspec list
  ```

By following these steps, you will have JupyterLab installed, running, and ready to use with your uploaded notebook file.
